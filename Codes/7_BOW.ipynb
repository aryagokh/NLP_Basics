```python
import pandas as pd
import numpy as np
import os
from sklearn.feature_extraction.text import CountVectorizer
```

```python
messages = pd.read_csv('./data/spam.csv', encoding='latin1')
messages.head()
```

```python
messages = messages[['v1', 'v2']]
messages.head()
```

```python
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()
```

```python
from tqdm import tqdm
corpus = []
for i in tqdm(range(0, len(messages))):
    review = re.sub('[^a-zA-Z]', ' ', messages['v2'][i]) #Regex
    review = review.lower() #Lowercase
    review = review.split() #List
    review = [stemmer.stem(word) for word in review if word not in set(stopwords.words('english'))] #Stemming and Stopwords
    review = ' '.join(review) #Again to sentence
    corpus.append(review)
```

```python
corpus[:5]
```

```python
# Creating BOW
from sklearn.feature_extraction.text import CountVectorizer
bow = CountVectorizer(max_features=2500)
```

```python
X = bow.fit_transform(corpus).toarray()
X
```

```python
X.shape, len(bow.vocabulary_)
```
